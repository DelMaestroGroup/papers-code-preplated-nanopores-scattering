```python
def μkey(μ):
    return f"μ = {μ:.1f} K"

def plot_SQ_exp(_ax):
    
    μ = [-47.0,-27.0,-19.0,-7.0]
    in_files = {'μ = -7.0 K':'f13.txt','μ = -19.0 K':'f9.txt', 'μ = -27.0 K':'f6.txt', 'μ = -47.0 K':'f4.txt'}
    μlabels = {'μ = -7.0 K':'13 mmol/g','μ = -19.0 K':'8.7 mmol/g', 'μ = -27.0 K':'5.9 mmol/g', 'μ = -47.0 K':'3.9 mmol/g'}

    for cμ in μ[::-1]:
        Q,S,ΔS = np.loadtxt(f'../data/{in_files[μkey(cμ)]}',skiprows=2,unpack=True)
        _ax.errorbar(Q,S,ΔS,**pore.get_props(pore_colors[μkey(cμ)], ls='--'), label=μlabels[μkey(cμ)])

  #  _ax.set_ylim(-50,580)
    _ax.set_xlim(1,3.52)
    _ax.set_xlabel('Q  (1/Å)')
    _ax.set_ylabel('S(Q,0)  (a.u.)')
    _ax.legend(loc=(0.0,0.625), ncol=1, fontsize=8,labelspacing=0.4 ); 
    
with plt.style.context(plot_style['sans']):
    fig,ax = plt.subplots(1,1,constrained_layout=True)
    dgutils.fonts.set_custom_font(font_path)
    plot_SQ_exp(ax)
    plt.savefig('../figures/exp_elastic_scattering.pdf')
```
